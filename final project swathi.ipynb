```python
class Restaurant:
    
    def __init__(self, name):
        self.name=name
        self.food={}
        self.food_Id=len(self.food)+1
        self.user_details={}
        self.ordered_item=[]

        
    def food_items(self):
        self.name = input("Enter the food name")
        self.quantity = int(input("Enter the quantity : "))
        self.price = float(input("Enter the price : "))
        self.stock = int(input("Enter the stock in kg"))
        self.discount = int(input("Enter discount value"))
        self.item = {"name":self.name , "quantity":self.quantity , "price":self.price , "stock":self.stock ,"discount":self.discount}
        self.food[self.food_id] = self.item
        
    def add_food_item(self):
        print("1. Confirm to add items \n 2. Exit")  
        k = input()    
        if k=="1":                
            while True:
                name = input("Enter the food name : ")
                quantity=float(input("Enter the quantity in values only : "))
                price=float(input("Enter the price in Rs only : "))
                discount=float(input("Enter the discount in Rs only : "))
                stock=float(input("Enter the available stock in values only : "))
                food_item={"Name":name,"Quantity":quantity,"Price":price,"Discount":discount,"Stock":stock}
                self.food_Id=len(self.food)+1
                self.food[self.food_Id]=food_item
                print("\n! Food Item added successfully !\n")
                print("Newly Added items :", self.food,"\n")
                break         
        elif k=="2":
            while True:
                main()
                break
        else:
            while True:
                print("something went wrong")
                break
    
    def edit_food_item(self):       
        x=int(input("Enter food id to update : "))
        if x in self.food.keys():
            print("1. Update Food Name. Update Quantity. Update Price. Update Discount. Update Stock ")
            y = input("Enter the number only : ")
            if y=="1":
                self.food[x]["Name"]=input("Updated food name : ")
                print("updated successfully: ")
            elif y=="2":
                self.food[x]["Quantity"]=float(input("Updated quantity in values only : "))
                print("updated successfully: ")
            elif y=="3":
                self.food[x]["Price"]=float(input("Updated price in Rs only : "))
                print("price updated successfully: ")
            elif y=="4":
                self.food[x]["Discount"]=float(input("Updated discount in Rs only : "))
                print(" updated successfully: ")
            elif y=="5":
                self.food[x]["Stock"]=float(input("Updated stock in values only : "))
                print("updated successfully:" )
            else:
                print("error")
        else:
            print("wrong food id")
     
    def view_food_item(self):
        print("List of food items : ")
        if len(self.food)!=0:    
            for i in self.food:
                print(f"Food Id : {i}")
                for j in self.food[i]:
                    print(j, ":", self.food[i][j])
                print()
        else:
            print("Invalid token!")
    
    def delete_food_item(self):        
            print("confirm to delete ")
            print("Enter the Food Id ")
            x = int(input())
            if x in self.food.keys():
                del self.food[x]
                print("Food item deleted successfully")
                print("Updated Food List\n",self.food)
            else:
                print("Incorrect Food Id")
    
    def user_register(self):        
            user_name=input("Enter your full name : ")
            phone_no=int(input("Enter your phone number : "))
            email=input("Enter your email id : ")
            password=input("Enter your password : ")
            address=input("Enter your address with PIN code ")
            self.user_details={"User Name":user_name,"Phone No.":phone_no,"Email_ID":email,"Password":password,
                               "Address":address}
            print(" Account is Created Successfully ")
            print("Welcome TO {self.name} Restaurant")
            print("User Details : ")
            for i in self.user_details:
                print(i, ":", self.user_details[i])
            
    def user_login(self):        
            while True:
                print(f"Welcome TO {self.name} Restaurant\n\nPlease Sign In")
                email=input("Enter Your email ID : ")
                pas=input("Enter Your Password : ")
                if len(self.user_details)!=0:
                    if email==self.user_details["Email_ID"]:
                        if pas==self.user_details["Password"]:
                            print("\n Login successfully ")
                            break
                        else:
                            print("\n Invalid Password \n")
                    else:
                        print("\n Invalid Email Id \n")
                else:
                    print("\n Account Doesn't Exist !\n\n!! Please create a new account \n")
                    self.user_register()
                    print(" Your account is created Successfully \n")

        
    def place_order(self):        
        menu=["1. Momos (8piece) [75Rs]", "2. Biryani (1 piece) [250Rs]","3.Rasmalai (1 piece) [40Rs]","4.Pastry (1piece) [60Rs]"]
        print("Menu")
        for i in menu:
            print(i)
        while True:
            print("\nEnter 1 to Place the Order\nEnter 2 to Exit ")
            x=input()
            if x=="1":
                print("Enter the Food number You want to order separated by comma")
                y=input().split(",")
                for i in y:
                    z=int(i)
                    if z<=len(menu):
                        self.ordered_item.append(menu[z-1])
                    else:
                        print("Food item is not available : ",z)
                print("Selected food items: ")
                for j in self.ordered_item:
                    print(j)
            elif x=="2":
                break
            else:
                print(" Invalid Number ")
              
                
    def ordered_history(self):
        print(" Previous orders : ")
        for i in self.ordered_item:
            print(i)
            
    def update_details(self):
        try:
            while True:
                print("Select below options to update your profile \n")
                print("1. Name\n2. Phone number\n3. Email ID\n4. Password\n5. Address\n6. Exit\n")
                x=input()
                if x=="1":
                    self.user_details["User Name"]=input("Enter new name : ")
                    print("\n!! Successfully Updated  !!\n")
                elif x=="2":
                    self.user_details["Phone No."]=int(input("Enter new phone number : "))
                    print("\n!! Successfully Updated !!\n")      
                elif x=="3":
                    self.user_details["Email_ID"]=input("Enter new email id : ")
                    print("\n!! Successfully Updated !!")
                    
                elif x=="4":
                    self.user_details["Password"]=input("Enter new password : ")
                    print(" Successfully Updated ")
                    
                elif x=="5":
                    self.user_details["Address"]=input("Enter new address with PIN code ")
                    print("\n!! Successfully Updated !!\n")
                    
                elif x=="6":
                    break
                else:
                    print(" Invalid Number ")
                    
                if x in ["1","2","3",'4',"5","6"]:
                    for i in self.user_details:
                        print(i, ":",self.user_details[i])
                else:
                    print("\nUser details updated :\n")      
        except Exception as e:
            print("\n!! Something went wrong please try again !!\n ")
        

            
def main():
    c = Restaurant("Swathi")    
    print("1. Admin\n2. User\n3. Exit\n")
    x=input()
    if x=="1":
        while True:
            print("1. Add New Food Items\n2. Edit Food Items\n3. View Food Items \n4. Delete Food Items\n5. Exit")
            y=input()
            if y=="1":
                c.add_food_item()
            elif y=="2":
                c.edit_food_item()
            elif y=="3":
                c.view_food_item()
            elif y=="4":
                c.delete_food_item()
            elif y=="5":
                main()
                break
            else:
                print("Invalid Number")
    elif x=="2":
        while True:
            print("\nEnter the Below Options\n")
            print("1. Register\n2. Login\n3. Exit\n")
            y=input()
            if y=="1":
                c.user_register()
            elif y=="2":
                c.user_login()
                while True:
                    print("\nEnter the Below Options\n")
                    print("1. Place New Order\n2. Check Order History\n3. Update Your Profile Details\n4. Exit")
                    z=input()
                    if z=="1":
                        c.place_order()
                    elif z=="2":
                        c.ordered_history()
                    elif z=="3":
                        c.update_details()
                    elif z=="4":
                        break
                    else:
                        print("invalid Number")            
            elif y=="3":
                main()
                break
            else:
                print("\nInvalid Number ")        
    elif x=="3": 
        while True:
            print("Thank you")
            break
    else:
        print("Invalid Number")
m = main()
```
